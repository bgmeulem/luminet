## Light ray curvature
Black holes curve spacetime around them. Light that comes closer than $\frac{3}{2}R_s=\frac{3}{2}2M=3M$ will enevitably be absorbed by the black hole, and is not visible to an external observer. For this reason, the distance of $3M$ is often called the "photon sphere", or the "shadow" of the black hole. For non-Swarzschild black holes, this value will differ, and may depend on the angle as well.

A ray of light that passes near a black hole, but stays at a distance larger than $2M$ will will deflect around the black hole:

![](../../assets/periastron.png)


In the example diagram above, a ray of light comes from a faraway light source, curves around a black hole, and reaches an observer somewhere on the right.

The nearest point between this photon orbit and the black hole is called the *periastron* and will be denoted with $P$. The name periastron is similar to perigee or perihelion, but uses the *-astron* suffix for a general stellar object, instead of *-gee* for earth (think of "geo") or *-helio* for sun. 

There is a direct between this periastron and the radial coordinate on the observer plane $b$ (already denoted in the diagram above):

$$
b^2 = \frac{P^3}{P - 2M}
$$

<div class="alert alert-info">

From this equation, you can already see:
- The larger the black hole mass, the smaller $b$ will be for the same periastron.
- A periastron equal to $2M$ will result in a division by zero, and a periastron smaller than $2M$ will lead to a negative $b^2$, or a complex number for the radial coordinate. These solutions are non-physical.

</div>




## Coordinates

Now that we mention coordinates on the observer plane, it's time to introduce our two coordinate systems, both in polar coordinates:
- The black hole coordinate system $(R, \alpha)$
- The observer coordinate system $(b, \alpha)$

To avoid confusion in the future:
- When talking about "radius", we always mean radius in the black hole reference frame ($R$)
- When talking about the radius in the observer frame ($b$), we will instead say **impact parameter**

It will be easy to think of the observer's frame as a CCD camera. A rectangular grid that is sensitive to light, and keeps track of where light rays end up. You may think cartesian coordinates may make more sense for a CCD camera pixel grid, but the relationship between the black hole reference frame and the photon orbits is much more simple to express in polar/spherical coordinates.

The relationship between the angles in both coordinate systems is trivial: they're the same, even after curving of the light path. That's why they are denoted with the same symbol too: $\alpha$. In the black hole reference frame, alpha spans the equatorial plane.

The relationship between the black hole frame radius $R$ and the radius of the observer frame $b$ is much, *much* less trivial:

$$
\frac{1}{R} = - \frac{Q - P + 2M}{4MP} + \frac{Q-P+6M}{4MP}{sn}^2\left( \frac{\gamma}{2}\sqrt{\frac{Q}{P}} + F(\zeta_\infty, k) \right)
$$
