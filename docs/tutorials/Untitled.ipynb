```python
from luminet.black_hole import BlackHole
import matplotlib.pyplot as plt
```

```python
bh = BlackHole(
    incl=1.4
)
```

```python
ax = None
highlight = 10
for ir in [50, 30, 10, 6]:
    color = "red" if ir == highlight else "white"
    ax = bh.plot_isoradials(
        direct_r=[ir],
        color=color,
        ax=ax
    )
plt.show()
```
