## The black hole accretion disk

Black holes have a disk of matter rotating around them: the accretion disk. This matter collides and shears as its being pulled inwards (being "accreted"). This friction produces light that can be seen. If you ever wondered how we can see black holes if they absorb light: this is how. All the light that is produced by either the accretion disk or by other light sources (faraway stars etc.) curves around the black hole. If it never comes closer than $2M$, it can curve towards our observer plane, our "camera".

## Calculating the relationship between $R$ and $P$

Given a fixed value for:
- radius $R$
- inclination $\theta_0$
- angle $\alpha$
- black hole mass (usually just $1$)

The previously mentioned equation gives us a relationship between $P$ and $R$. Let's consider an example:

```python
mass = 1
radius = 20*mass
inclination = 1.4  # slightly less than perfectly edge-on
angle = 3
```

We can re-arrange the previous equation so that it equals zero when the value for $P$ matches the value for $R$:

$$
0 = \frac{1}{R} + \frac{Q - P + 2M}{4MP} - \frac{Q-P+6M}{4MP}{sn}^2\left( \frac{\gamma}{2}\sqrt{\frac{Q}{P}} + F(\zeta_\infty, k) \right)
$$

If we simply guess a bunch of values for $P$, we see:

```python
import luminet.black_hole_math as bh_math
import numpy as np

guess_values_for_periastron = np.linspace(6,30,30)

result = [
    bh_math.perigee_optimization_function(
        ir_radius=radius,
        incl=inclination,
        ir_angle=angle,
        bh_mass=mass,
        p=P
    )
    for P in guess_values_for_periastron
]
```

```python
import matplotlib.pyplot as plt
plt.plot(guess_values_for_periastron, result)
```

Only the value for $P$ that makes the function equal $0$ is a correct solution. let's find out which one of our guesses was closest:

```python
ind_closest_P_to_correct = np.argmin([abs(e) for e in result])
closest_P_to_correct = guess_values_for_periastron[ind_closest_P_to_correct]
```

```python
closest_P_to_correct
```

```python
import matplotlib.pyplot as plt
plt.plot(guess_values_for_periastron, result)
plt.axvline(closest_P_to_correct, c='red')
plt.axhline(result[ind_closest_P_to_correct], c='red')
```

We see that one of our guesses came really close to a value of $0$. This was of course just guesswork. A more robust way is to apply an actual solver to this problem. That's what this repository helps with. It simply applies a numpy solver to the equation above.

## Calculating isoradial lines

Isoradial lines are lines of equal distance to the black hole, as the appear on the observer frame. In other words, they are equidistance rings from the black hole, as they appear after being curved around the black hole. Equidistant rings is exactly what the accretion disk is (at least, if it is thin enough). So, if we can visualize the shape of these lines of equal radius, we can visualize the shape of the black hole's accretion disk.

Let's take an inclination slightly above the accretion disk for the most extreme effect. Note that the inclination is counted from the top view (see the coordinate system above), so a topview would be inclination $0$, and a perfect edge-on inclination would be $\pi/2$

```python
inclination = 1.4  # slightly less than perfectly edge-on
```

```python
b = bh_math.solve_for_impact_parameter(
    radius=10,
    incl=inclination,
    alpha=0,
    bh_mass=1,
)
b
```

If we now monotonically increase the angle from $0$ to $2\pi$, we can visualize what a ring looks like given the previous coordinates

```python
angular_resolution = 100
```

```python
impact_parameters = []
angles = bh_math.np.linspace(0, bh_math.np.pi*2, angular_resolution)
for angle in angles:
    b = bh_math.solve_for_impact_parameter(
        radius=10,
        incl=inclination,
        alpha=angle,
        bh_mass=1,
    )
    impact_parameters.append(b)
```

```python
import matplotlib.pyplot as plt
plt.plot(impact_parameters)
```

Here we see how the radius on the observer frame varies as we follow a ring of equal distance in the black hole reference frame. Remember that all these photons come from the same distance $R=10M$.

In other words, if we plot this in polar coordinates, we can see what the ring of $10M$ looks like on our observer frame!

```python
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(
    angles,
    impact_parameters,
)
ax.set_theta_zero_location('S')
```

This is what the `Isoradial` class automates:

```python
from luminet.isoradial import Isoradial
```

```python
ir = Isoradial(
    radius = 10,
    incl=inclination,
    bh_mass = 1,
    angular_resolution=angular_resolution
)
```

```python
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.set_theta_zero_location('S')
ax = ir.plot(ax=ax)
```
